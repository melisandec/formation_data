```python
df.drop(columns = ['incident_id', 'city_or_county', 'address'], inplace = True)
df.head()
```

```python
data = df.copy()
```

```python
from datetime import datetime 

data['date'] = pd.to_datetime(data['date']) # convert date to datetime format
data['year'] = data['date'].dt.year # extract year
data['month'] = data['date'].dt.month # extract month
data['day'] = data['date'].dt.day # extract day
data['day_of_week'] = data['date'].dt.dayofweek # extract day of week
```

```python
data.head()
```

### Mapping

```python
# Replace numeric day of week values with text

data['day_of_week'] = data['day_of_week'].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
data.head()
```

## Dealing with missing values

```python
# Création d'un DataFrame avec 10 lignes et 5 colonnes
data = {'Nom': ['Alice', 'Bob', 'Charlie', np.nan, 'Emma', 'Frank', 'Grace', 'Hank', 'Ivy', 'Jack'],
        'Âge': [25, 30, 35, 28, 22, np.nan, 32, 29, np.nan, np.nan],
        'Ville': ['Paris', 'New York', 'Londres', 'Paris', 'New York', 'Londres', 'Paris', 'New York', 'Londres', np.nan],
        'Salaire': [60000, 75000, 80000, 50000, 70000, 90000, 55000, 72000, 68000, 80000],
        'Score': [75, 85, 90, np.nan, 95, 88, 92, 78, 87, 94]}

df = pd.DataFrame(data)
df
```
